```python
from IPython.core.interactiveshell import InteractiveShell
from stereomolgraph import StereoCondensedReactionGraph
from stereomolgraph.ipython import View2D
from stereomolgraph.stereodescriptors import PlanarBond, Tetrahedral

View2D.show_atom_numbers = True
View2D.generate_bond_orders = True

InteractiveShell.ast_node_interactivity = "all"
```

```python
scrg1 = StereoCondensedReactionGraph()

for id, atom_type in enumerate([6, 6, 6, 6, 6, 6, 1, 1, 1, 1, 1, 1, 1, 1]):
    scrg1.add_atom(id, atom_type)

for bond in (
    (0, 1),
    (1, 2),
    (2, 3),
    (3, 4),
    (4, 5),
    (0, 6),
    (0, 7),
    (1, 8),
    (2, 9),
    (3, 10),
    (4, 11),
    (5, 12),
    (5, 13),
):
    scrg1.add_bond(*bond)

scrg1.add_formed_bond(0, 5)

scrg1.set_bond_stereo_change(broken=PlanarBond((6, 7, 0, 1, 8, 2), 0))
scrg1.set_bond_stereo_change(broken=PlanarBond((1, 9, 2, 3, 4, 10), 0))
scrg1.set_bond_stereo_change(broken=PlanarBond((3, 11, 4, 5, 12, 13), 0))

scrg1.set_bond_stereo_change(formed=PlanarBond((0, 8, 1, 2, 3, 9), 0))
scrg1.set_bond_stereo_change(formed=PlanarBond((11, 5, 4, 3, 10, 2), 0))

scrg2 = scrg1.copy()

scrg1.set_atom_stereo_change(formed=Tetrahedral((0, 1, 7, 6, 5), 1))
scrg1.set_atom_stereo_change(formed=Tetrahedral((5, 4, 12, 13, 0), 1))

scrg2.set_atom_stereo_change(formed=Tetrahedral((0, 1, 7, 6, 5), 1))
scrg2.set_atom_stereo_change(formed=Tetrahedral((5, 4, 13, 12, 0), 1))

scrg1.reactant() == scrg2.reactant()
scrg1.product() == scrg2.product()
scrg1 == scrg2

scrg1.reactant()
scrg1
scrg1.product()
scrg2
scrg2.product()
```
