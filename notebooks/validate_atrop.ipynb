```python
import rdkit
from rdkit import Chem

from stereomolgraph.stereodescriptors import AtropBond
from stereomolgraph import StereoMolGraph
from stereomolgraph.ipython import View2D

View2D.show_h = True

rdkit.__version__
```

```python
expected_chirality = {23: "Chiral",
55: "Chiral",
57: "Chiral",
72: "Chiral",
73: "Chiral",
86: "Chiral",
158: "Chiral"}
```

```python
# Read SMILES from text file
def load_smiles_from_txt(filepath: str) -> list[str]:
    """Load SMILES strings from a text file"""
    with open(filepath, 'r') as f:
        smiles_list = [line.strip() for line in f if line.strip()]
    print(f"Loaded {len(smiles_list)} SMILES from {filepath}")
    return smiles_list

# Load SMILES from file
smiles_list = load_smiles_from_txt('../data/smiles.txt')
```

```python
smiles = [smiles_list[i-1].split()[0] for i in expected_chirality]
```

```python
def atropoisomer_from_smiles(smiles: str) -> StereoMolGraph:
    """Create a StereoMolGraph from a SMILES string"""
    mol = Chem.AddHs(Chem.MolFromSmiles(smiles))
    if mol is None:
        raise ValueError(f"Invalid SMILES: {smiles}")
    return StereoMolGraph.from_rdmol(mol)
```

```python
smg1 = atropoisomer_from_smiles(smiles[0])
smg1
```

```python
smg1.set_bond_stereo(AtropBond((3,15,4,5,6,11), 1))
smg1 == smg1.enantiomer()
```
