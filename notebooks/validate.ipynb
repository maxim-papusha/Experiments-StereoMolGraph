```python
# Read SMILES from text file
def load_smiles_from_txt(filepath: str) -> list[str]:
    """Load SMILES strings from a text file"""
    with open(filepath, 'r') as f:
        smiles_list = [line.strip() for line in f if line.strip()]
    print(f"Loaded {len(smiles_list)} SMILES from {filepath}")
    return smiles_list

# Load SMILES from file
smiles_list = load_smiles_from_txt('../data/smiles.txt')
```

```python
def smiles_to_mol(smiles: str) -> Chem.Mol:
    return Chem.MolFromSmiles(smiles)

def mol_to_inchi(mol: Chem.Mol) -> str:
    molblock = Chem.rdmolfiles.MolToMolBlock(mol)
    return Chem.inchi.MolBlockToInchi(molblock)

def mol_to_canon_smiles(mol: Chem.Mol) -> str:
    return Chem.MolToSmiles(mol, canonical=True)
```

```python
def generate_renumbered_molecules(mol: Chem.Mol, n: int = RENUMBER_TIMES
                                  ) -> Iterator[Chem.Mol]:
    """Generate n molecules with renumbered atoms from an rdmol"""
    for _ in range(n):
        # Generate random atom order
        num_atoms = mol.GetNumAtoms()
        new_order = list(range(num_atoms))
        random.shuffle(new_order)
        # Create renumbered molecule
        renumbered_mol = Chem.rdmolops.RenumberAtoms(mol, new_order)
        yield renumbered_mol
```

```python
def inchi_identety(mol: Chem.Mol, n: int = RENUMBER_TIMES):
    unique_strings: set[str] = set()
    for renumbered_mol in generate_renumbered_molecules(mol, n):
        unique_strings.add(mol_to_inchi(renumbered_mol))
    if len(unique_strings) == 1:
        return "✓"
    else:
        return "✗"
```

```python
def smiles_identety(mol: Chem.Mol, n: int = RENUMBER_TIMES):
    unique_strings: set[str] = set()
    for renumbered_mol in generate_renumbered_molecules(mol, n):
        unique_strings.add(mol_to_canon_smiles(renumbered_mol))
    if len(unique_strings) == 1:
        return "✓"
    else:
        return "✗"
```

```python
def smg_identity(mol: Chem.Mol, n: int = RENUMBER_TIMES):
    mol = Chem.AddHs(mol)
    initial_smg = StereoMolGraph.from_rdmol(mol, stereo_complete=True)
    for renumbered_mol in generate_renumbered_molecules(mol, n):
        renumbered_smg = StereoMolGraph.from_rdmol(renumbered_mol, stereo_complete=True)
        if renumbered_smg != initial_smg:
            return "✗"
    return "✓"
```

```python
def generate_enantiomer(mol: Chem.Mol) -> Chem.Mol:
    """Generate enantiomer by inverting all chiral tetrahedral stereocenters"""
    # Create a copy to avoid modifying the original
    enantiomer = Chem.RWMol(mol)
    
    for atom in enantiomer.GetAtoms():
        if atom.GetChiralTag() != Chem.ChiralType.CHI_UNSPECIFIED:
            # Invert the chirality
            if atom.GetChiralTag() == Chem.ChiralType.CHI_TETRAHEDRAL_CW:
                atom.SetChiralTag(Chem.ChiralType.CHI_TETRAHEDRAL_CCW)
            elif atom.GetChiralTag() == Chem.ChiralType.CHI_TETRAHEDRAL_CCW:
                atom.SetChiralTag(Chem.ChiralType.CHI_TETRAHEDRAL_CW)
    
    enantiomer = enantiomer.GetMol()
    return enantiomer
```

```python
def chython_identity(mol: Chem.Mol, n: int = RENUMBER_TIMES):
    """Check if chython can identify renumbered molecules as identical"""
    try:
        # Convert initial molecule to chython via SMILES
        initial_smiles = Chem.MolToSmiles(mol)
        initial_chython = chython.smiles(initial_smiles)
        # Check if all renumbered versions are recognized as identical by chython
        for renumbered_mol in generate_renumbered_molecules(mol, n):
            renumbered_smiles = Chem.MolToSmiles(renumbered_mol, doRandom=True, canonical=False)
            renumbered_chython = chython.smiles(renumbered_smiles)
            if renumbered_chython != initial_chython:
                return "✗"
        return "✓"
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
def inchi_enantiomer_equal(mol: Chem.Mol):
    """Check if molecule's InChI equals its enantiomer's InChI"""
    try:
        enantiomer = generate_enantiomer(mol)
        mol_inchi = mol_to_inchi(mol)
        enantio_inchi = mol_to_inchi(enantiomer)
        if mol_inchi == enantio_inchi:
            return "Achiral"
        else:
            return "Chiral"
    except Exception as e:
        return f"Error: {str(e)}"

def smiles_enantiomer_equal(mol: Chem.Mol):
    """Check if molecule's canonical SMILES equals its enantiomer's canonical SMILES"""
    try:
        enantiomer = generate_enantiomer(mol)
        mol_smiles = mol_to_canon_smiles(mol)
        enantio_smiles = mol_to_canon_smiles(enantiomer)
        if mol_smiles == enantio_smiles:
            return "Achiral"
        else:
            return "Chiral"
    except Exception as e:
        return f"Error: {str(e)}"

def smg_enantiomer_equal(mol: Chem.Mol):
    """Check if molecule's SMG equals its enantiomer's SMG"""
    try:
        mol = Chem.AddHs(mol)
        smg = StereoMolGraph.from_rdmol(mol, stereo_complete=True)
        enantiomer_smg = smg.enantiomer()
        if smg == enantiomer_smg:
            return "Achiral"
        else:
            return "Chiral"
    except Exception as e:
        return f"Error: {str(e)}"

def chython_enantiomer_equal(mol: Chem.Mol):
    """Check if molecule's Chython representation equals its enantiomer's"""
    try:
        enantiomer = generate_enantiomer(mol)
        mol_smiles = Chem.MolToSmiles(mol)
        enantiomer_smiles = Chem.MolToSmiles(enantiomer, canonical=False)
        mol_chython = chython.smiles(mol_smiles)
        enantiomer_chython = chython.smiles(enantiomer_smiles)
        if mol_chython == enantiomer_chython:
            return "Achiral"
        else:
            return "Chiral"
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
def validate_smiles_list(smiles_list: list[str]) -> tuple[pd.DataFrame, pd.DataFrame]:
    """
    Validates a list of SMILES strings by checking if they are maintained
    under atom renumbering for InChI, SMILES, SMG, and Chython representations.
    Also checks if each molecule is equal to its enantiomer.
    
    Returns two DataFrames:
    1. Identity checks under renumbering
    2. Enantiomer equality checks
    """

    identity_results = []
    enantiomer_results = []
    
    for i, smiles_input in enumerate(smiles_list, 1):
        if i in ids_to_skip:
            continue
        smiles_input, *_ = smiles_input.split()
        mol = smiles_to_mol(smiles_input)
        if mol is None:
            continue

        # skip if with isotopes
        if any(atom.GetIsotope() != 0 for atom in mol.GetAtoms()):
            continue
        
        identity_results.append({
            
            'ID': f"VS{i:03d}",
            'RDKit depiction': mol,
            'StereoMolGraph': smg_identity(mol),
            'InChI': inchi_identety(mol),
            'RDKit SMILES': smiles_identety(mol),
            'chython': chython_identity(mol),
            'Input SMILES': smiles_input
        })
        
        enantiomer_results.append({
            'ID': f"VS{i:03d}",
            'Mol': mol,
            'manual': manual_assignment[f"VS{i:03d}"],
            'StereoMolGraph': smg_enantiomer_equal(mol),
            'InChI': inchi_enantiomer_equal(mol),
            'RDKit SMILES': smiles_enantiomer_equal(mol),
            'chython': chython_enantiomer_equal(mol),
            'Input SMILES': smiles_input
        })
    
    df_identity = pd.DataFrame(identity_results)
    df_enantiomer = pd.DataFrame(enantiomer_results)

    return df_identity , df_enantiomer
```

```python
# Compare each method against manual assignment
def compare_with_manual(df: pd.DataFrame, method_col: str) -> dict:
    """Compare a method's results with manual assignment"""
    # Filter out rows where method returned an error
    valid_df = df[~df[method_col].str.contains('Error', na=False)]
    
    total = len(valid_df)
    mismatches = valid_df[valid_df['manual'] != valid_df[method_col]]
    num_mismatches = len(mismatches)
    percentage = (num_mismatches / total * 100) if total > 0 else 0
    
    return {
        'total': total,
        'mismatches': num_mismatches,
        'percentage': percentage,
        'mismatch_ids': mismatches['ID'].tolist()
    }

# Analyze each method
methods = ['StereoMolGraph', 'InChI', 'RDKit SMILES', 'chython']
results = {}

for method in methods:
    results[method] = compare_with_manual(df_enantiomer, method)

# Print summary
print("="*70)
print("COMPARISON WITH MANUAL ASSIGNMENT")
print("="*70)
print()

for method in methods:
    r = results[method]
    print(f"{method}:")
    print(f"  Total valid comparisons: {r['total']}")
    print(f"  Mismatches: {r['mismatches']}")
    print(f"  Accuracy: {100 - r['percentage']:.2f}%")
    print(f"  Error rate: {r['percentage']:.2f}%")
    print()

# Print detailed mismatch lists
print()
print("="*70)
print("DETAILED MISMATCH LISTS")
print("="*70)
print()

for method in methods:
    r = results[method]
    print(f"{method} mismatches ({r['mismatches']} cases):")
    if r['mismatch_ids']:
        print(f"  {', '.join(r['mismatch_ids'])}")
    else:
        print("  None")
    print()
```
