```python
from IPython.core.interactiveshell import InteractiveShell
from stereomolgraph import StereoCondensedReactionGraph
from stereomolgraph.ipython import View2D
from stereomolgraph.stereodescriptors import Tetrahedral

View2D.show_atom_numbers = True
View2D.generate_bond_orders = True

InteractiveShell.ast_node_interactivity = "all"
```

```python
scrg1 = StereoCondensedReactionGraph()

for id, a in enumerate(
    ("O", "C", "H", "H", "C", "H", "H", "C", "H", "H", "F")
):
    scrg1.add_atom(id, a)

for a1, a2 in [
    (0, 1),
    (1, 2),
    (1, 3),
    (1, 4),
    (4, 5),
    (4, 6),
    (4, 7),
    (7, 8),
    (7, 9),
    (7, 10),
]:
    scrg1.add_bond(a1, a2)

for tup in [(1, 0, 2, 3, 4), (4, 6, 7, 5, 1), (7, 8, 9, 4, 10)]:
    scrg1.set_atom_stereo(Tetrahedral(tup, 1))

scrg2 = scrg1.copy()

scrg1.add_formed_bond(0, 8)
scrg1.add_broken_bond(8, 7)

scrg2.add_formed_bond(0, 9)
scrg2.add_broken_bond(9, 7)

are_equal = scrg1 == scrg2
are_enantiomers = scrg1.enantiomer() == scrg2

print(f"The TS {'' if are_equal else 'not'} are equal !")
print(
    f"The TS are {'' if are_enantiomers else 'not'}enantiomes of eachother !"
)

scrg1
scrg2
```
